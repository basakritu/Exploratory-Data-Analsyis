```python
country_code = pd.read_excel('Country-Code.xlsx')
country_code.head()
```

```python
final_df = pd.merge(df, country_code, on = 'Country Code', how= 'left')
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
values = final_df.Country.value_counts().values
```

```python
#pie chart top 3 countries uses zomato

plt.pie(values[:3],labels = country_names[:3], autopct='%1.2f%%')
```

observation: India uses most zomato, then usa and uk

```python
rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
rating
```

```python
sns.barplot(rating['Aggregate rating'],  rating['Rating count'], hue = rating['Rating color'],
            palette = ['white', 'red','orange', 'yellow', 'green', 'green'])
# plt.figure(size = (12,6))
```

```python
sns.countplot(x = 'Rating color', data=rating, palette = ['white', 'red','orange', 'yellow', 'green', 'green'])
```
